```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

>Because of wrong, duplicate columns name, nan values, so it needs to rename the data columns used by Symbol columns with no duplicate.
Def for rename data frame columns, drop unnecessary columns and rows.

```python
def createDf(df):
    column_name = list(df['Symbol'].dropna())
    column_name.insert(0,"Date")

    new_df = df.drop(['Name', 'Symbol', 'RIC'], axis=1)
    new_df = new_df.drop([0,1,2])
    new_df = new_df.set_axis(column_name, axis=1).set_index('Date')

    return new_df
```

```python
# Import dataframe

df_price = createDf(pd.read_excel("Sri Lanka - price - market cap.xlsx", sheet_name="Price"))
df_marketcap = createDf(pd.read_excel("Sri Lanka - price - market cap.xlsx", sheet_name="Market cap"))
df_icb =  pd.read_excel("Sri Lanka ICB.xlsx")
```

> Because of the similarity between 2a and 2b, so we create the functions that can use for two

```python
# def for create data
def data2ab(day):
    #Prepare data
    df = df_price.loc[[day, '2021-09-24']]
    df = df.transpose().dropna() # Tranpose the data to caculate the return
    df['Rate of return'] = (df['2021-09-24'] - df[day])/df[day]

    # The types of 'Rate of return' is object => turn to float to use def nlargest
    df['Rate of return'] = df['Rate of return'].astype(float)
    return df

# def for plot data
def plotdata(df, palette):
    plt.subplots(figsize=(12, 8))
    sns.barplot(data= df, y=df.index, x=df['Rate of return'], palette=palette)
    plt.show()
```

<h2>The top 10 stocks have the highest rate of return compare to the first day of the month (September 1, 2021)

```python
#Prepare data for 2a
df_price2a = data2ab('2021-09-01').nlargest(10, 'Rate of return')
df_price2a
```

```python
# Plot 2a
plotdata(df_price2a, 'flare')
```

<h2>The top 10 stocks have the lowest rate of return compare to the first day of the year (January 1, 2021)

```python
#Prepare data for 2b
df_price2b = data2ab('2021-01-01').nsmallest(10, 'Rate of return')
df_price2b
```

```python
# Plot 2b
plotdata(df_price2b, 'viridis')
```

<h2>Stocks that are priced above the 52-week mark

```python
# Get data of 52 weeks
df_price_52week = df_price.loc['2020-09-23': '2021-09-23'].transpose().dropna()

# Get the max value of 52 weeks
df_price_52week['Max_52week'] = df_price_52week.max(axis=1)
df_price2c = df_price_52week[['Max_52week']].copy()

# Join data of max and 24/9
df_price2c['Price_24_9'] = df_price.loc['2021-09-24'].transpose().dropna()

# Create column 'Condition' with return True if price at present > max price in 52 week
df_price2c['Condition'] = df_price2c['Price_24_9'] > df_price2c['Max_52week']
df_price2c = df_price2c.loc[(df_price2c['Condition'] == True)]

df_price2c
```

```python
df_price2c.plot(kind= 'bar', rot=0, figsize=(12,8))
plt.show()
```

<h2>Five stocks have the largest market cap in each industry at present

This block below will be use for 2d and 2e

```python
# Prepare data from sheet marketcap
df_marketcap2de = df_marketcap.transpose().reset_index().rename(columns={'index': 'Symbol'})

# Merge data market cap and data icb to get industry name, key is Symbol, then get necessary columns
day24 = df_marketcap2de.columns[-1]
df_marketcap2de = df_marketcap2de.merge(df_icb, on='Symbol')
df_marketcap2de = df_marketcap2de[['Symbol', 'ICB INDUSTRY NAME', day24]].dropna().reset_index(drop=True)

# Rename the 24-9-2021 market cap columns to Value
column_name = ['Symbol', 'ICB INDUSTRY NAME', 'Value']
df_marketcap2de = df_marketcap2de.set_axis(column_name, axis=1)
df_marketcap2de['Value'] = df_marketcap2de['Value'].astype(float)
```

```python
# Get top five stocks have largest market cap in each industry 
df_marketcap2d = df_marketcap2de.groupby(['ICB INDUSTRY NAME']).apply(lambda x: x.nlargest(5, 'Value')).reset_index(drop=True)
df_marketcap2d
```

```python
# create list of industry data
dict_industry = df_marketcap2d['ICB INDUSTRY NAME'].value_counts().to_dict()
lst_industry = list(dict_industry.keys())

lst_top5 = []
for i in lst_industry:
    lst_top5.append(df_marketcap2d[df_marketcap2d['ICB INDUSTRY NAME'] == i].reset_index(drop=True))

# def for plot
def plot2d(df):
    plt.title(df.iloc[0,1])
    sns.barplot(x='Symbol', y='Value', data=df)

# create axes and plot
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(25, 20))  

for a in range(3):
    for b in range(4):
        try:
            plt.sca(axes[a,b])
            plot2d(lst_top5[a+b+3*a])
        except:
            pass
        
plt.show()
```

<h2>The overall picture of the total market capitalization of industries at the present time. 

```python
df_marketcap2e = df_marketcap2de.groupby(['ICB INDUSTRY NAME']).sum()
df_marketcap2e
```

```python
#create pie chart
colors = sns.color_palette('Paired')
plt.subplots(figsize=(12, 12))
plt.title('THE TOTAL MARKET CAP')
plt.pie(df_marketcap2e['Value'], labels = df_marketcap2e.index, autopct="%.1f%%", colors = colors)
plt.show()
```
